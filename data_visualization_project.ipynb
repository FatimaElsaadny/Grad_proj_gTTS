```python
#from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
names = df.columns
d = scaler.fit_transform(df)
scaled_df = pd.DataFrame(d, columns=names)
scaled_df
```

```python
import seaborn as sns
matrix_corr = scaled_df.corr()
print(matrix_corr)
sns.heatmap(matrix_corr, vmax=1, square=True)
# matrix_corr = scaled_df.corr()
# matrix_corr['Smoking'].sort_values(ascending = False)
```

```python
g = sns.PairGrid(scaled_df, vars=['Diabetic', 'SkinCancer', 'KidneyDisease', 'Stroke'],
                 hue='HeartDisease', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```
